```python
import requests
import pandas as pd
```

```python
#總頁數

url = "https://ecshweb.pchome.com.tw/search/v3.3/all/results?q=滑鼠&page={}&sort=sale/dc"

payload={}
headers = {
  'Cookie': 'U=39b74b049dd63873188c1d502e74ff60e1770830'
}

response = requests.request("GET", url.format(1), headers=headers, data=payload)

res_dict = response.json()
total_page = res_dict['totalPage']

total_page
```

```python
data = []

for i in range(1,total_page)[:10]:
   res = requests.get(url.format(i))
   res_dict = res.json()

   for prod in res_dict['prods']:
          print(prod['name'], prod['price'])
          data.append({
                  'name':prod['name'],
                  'price':prod['price']
                       })
```

```python
df =pd.DataFrame(data)
print(df)
```
